```python
df['Total yearly emissions']=df[['Buildings', 'Industry',
       'Land-use change and forestry', 'Other fuel combustion', 'Transport',
       'Manufacturing and construction', 'Fugitive emissions',
       'Electricity and heat']].sum(axis=1)
```

```python
I#Total Carbon di oxide emissions in CO2 equivalents
plt.rcParams['figure.figsize'] = [10,16]
plt.plot(df['Year'][0:30],df['Total yearly emissions'][0:30],label="Argentina")
plt.plot(df['Year'][0:30],df['Total yearly emissions'][30:60],label="Brazil")
plt.plot(df['Year'][0:30],df['Total yearly emissions'][60:90],label="Chile")
plt.plot(df['Year'][0:30],df['Total yearly emissions'][90:120],label="India")

plt.title("Total Carbon di oxide emissions in CO2 equivalents")
plt.legend()
plt.show()
```

Conclusions : 
* The total carbondioxide emissions by India has a highly positive trend and a high value due to its population
* The carbondioxide emissions by Brazil has decreased significantly after 2010.
* The carbondioxide emissions by Argentina and Chile are almost constant


```python
 #Contribution of different sectors to Argentina's CO2 emissions:
plt.rcParams['figure.figsize'] = [10,6]
for i in df.columns[3:-1]:
  plt.plot(df['Year'][0:30],df[i][0:30],label=i)

plt.title('Sector wise carbon di oxide emissions in CO2 equivalents, Argentina')
plt.xlabel('Years')
plt.ylabel('Billion CO2 eq')
plt.legend()
plt.show()
```

Conclusions:
* Land use and forestry sector used to be the largest contributor but it has a sognificant drop after 2010
* Electricity and heat is the leading contributor now, followed by Transport. There is a downward trend in the contribution by electicity and heat sector lately, however the contribution by transport sector is almost stagnant.

Targets:The governments target of reaching emissions levels of 313 MtCO2e by 2030 (excluding LULUCF) is not compatible with a 1.5C pathway.

The contribution by the land use and forestry sector in Brazil decreased drastically at around 2010.

The contribution in other three countries has more or less stayed stagnant this century, and in India the forests have started becoming a source from being a sink in the early 2000s

Argentina: The 2007 National Forests law set ambitious LULUCF goals, outlining protection for native forests and biodiverse areas against deforestation, however the enforcement was not fully funded. In 2021,the president stated that there will be a stronger emphasis on eradicating illegal deforestation, but has not acted on this.

Brazil: Brazil has set ambitious targets to reduce deforestation, aiming to reduce deforestation by 80% by 2020 relative to the reference period 1996-2005. However, despite a historically low deforestation rate in 2012,deforestation has been gradually increasing since, and spiked after the covid 19 pandemic

Chile: Chile's forest sinks are vulnerable to extreme weather events.The national strategy for climate change and Vegetation resources was established in 2013 along with regulations and instruments granting incenives for forest owners to preserve or create new forests. These sinks are very necessary for Chile to achieve its target of carbon neutrality by 2050.

India: India has pledged to create an additional carbon sink of 2.5-3.0 billion tonnes of CO2e by 2030. India's current policies of expansion of coal mining and clearing of forested land for infrastructure and industry developments are not in line with this kind of sink creation.
